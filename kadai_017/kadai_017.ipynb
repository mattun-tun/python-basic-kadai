```python
class Human:
  def __init__(self,name, age):
    self.name = name
    self.age = age
  def check_adult(self):
    if self.age >= 20:
      print(self.name + 'is adult.')
    else:
      print(self.name + ' is not adult.')
humans = [Human('太郎',25),Human('二郎',20),Human('三郎',18)]
for human in humans:
    human.check_adult()
```
